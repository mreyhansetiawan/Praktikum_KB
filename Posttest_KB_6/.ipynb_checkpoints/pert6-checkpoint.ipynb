```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
```

```python
iris = datasets.load_iris()
```

```python
dataset = pd.DataFrame(iris.data, columns=iris.feature_names)
dataset['Target'] = iris.target
```

```python
dataset.head(5)
```

```python
model = KMeans(n_clusters=5)
```

```python
points = dataset[['sepal length (cm)', 'petal length (cm)']]
```

```python
model.fit(points)
```

```python
labels = model.labels_
```

```python
labels
```

```python
dataset_KMeans = pd.DataFrame({'labels' : labels, 'species' : dataset.Target})
```

```python
crosstab = pd.crosstab(dataset_KMeans['labels'], dataset_KMeans['species'])
```

```python
crosstab
```

```python
X = points['sepal length (cm)']
Y = points['petal length (cm)']
```

```python
centroids = model.cluster_centers_
```

```python
centroids_X = centroids[:,0]
centroids_Y = centroids[:,1]
```

```python
plt.figure(figsize=(8,8))
plt.scatter(X,Y, alpha=0.5,  c=labels)
plt.scatter(centroids_X, centroids_Y, marker="D", s=5, label='Centroid')
plt.xlabel('sepal length (cm)')
plt.ylabel('petal length (cm)')
plt.legend()
plt.show()
```

```python
kluster_range = range(1, 9)
inertias = []

for k in kluster_range:
    model = KMeans(n_clusters=k)
    model.fit(points)
    inertias.append(model.inertia_)
```

```python
plt.figure(figsize=(8,8))
plt.plot(kluster_range, inertias,'-o')
plt.xlabel('jumlah klaster')
plt.ylabel('inertia')
plt.show()
```

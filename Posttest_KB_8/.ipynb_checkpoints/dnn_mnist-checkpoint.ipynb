```python
import tensorflow as tf
import tensorflow.keras.datasets as tfds
import matplotlib.pyplot as plt
import numpy as np
```

```python
(x_train, y_train), (x_test, y_test) = tfds.mnist.load_data()

print(f"panjang x train : {len(x_train)} gambar",
        f"panjang y train : {len(y_train)} gambar",
        f"panjang y test  : {len(y_test)} gambar",
        f"panjang y test  : {len(y_test)} gambar",
        sep="\n"
)
```

```python
num_class = 10
y_train = tf.keras.utils.to_categorical(y_train, num_class)
y_test = tf.keras.utils.to_categorical(y_test, num_class)
```

```python
plt.imshow(x_train[32])
```

```python
x_train = x_train.reshape(60000, 784)
x_test = x_test.reshape(10000, 784)
```

```python
x_train = x_train.astype("float32")/255
x_test = x_test.astype("float32")/255
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(32, activation="relu", input_dim=784),
    tf.keras.layers.Dense(num_class, activation="softmax")
])
model.summary()
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(),
    loss=tf.keras.losses.categorical_crossentropy,
    metrics=["accuracy"]
)
```

```python
history = model.fit(
    x = x_train, y = y_train, 
    validation_data=(x_test, y_test)
)
```

```python
history = model.fit(
    x = x_train, y = y_train, 
    validation_data=(x_test, y_test),
    epochs=10,
    verbose=2,
    steps_per_epoch=500
)
```

```python
plt.figure(1, figsize=(5,5))
plt.plot(range(len(history.history["loss"])), 
        history.history["loss"], 
        label="training")
plt.plot(range(len(history.history["loss"])), 
        history.history["loss"], 
        label="validation")
plt.legend()
plt.show()
```

```python
model.save("model/model1.h5")
```

```python
saved_model = tf.keras.models.load_model("model/model1.h5")
```

```python
saved_model.evaluate(x_test, y_test, batch_size=32)
```

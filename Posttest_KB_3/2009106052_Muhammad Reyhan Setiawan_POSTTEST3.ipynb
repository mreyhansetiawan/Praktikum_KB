# Membuat 4 Tipe Grafik Visualisasi Data

# 1. Grafik Batang

Bar Chart atau Grafik Batang sering digunakan untuk membandingkan kategori. Pada Grafik batang di bawah ini akan membandingkan banyaknya ulasan per Kategori App

```python
x = df["Category"]
y = df["Reviews"]
```

```python
plt.bar(x, y, color="blue", )

plt.title("Perbandingan banyaknya Ulasan per Kategori") 
plt.xlabel("Kategori")
plt.xticks(rotation = 90)
plt.ylabel("Ulasan")
plt.show()
```

Hasil pada grafik batang di atas menunjukkan App kategori Social merupakan kategori dengan ulasan paling banyak

# 2. Grafik Pie

Pie Chart atau Grafik Pie biasanya digunakan untuk data kategorikal. Pada Grafik Pie di bawah ini akan melihat distribusi banyaknya App per Kategori

```python
df["Category"].value_counts()
```

```python
df = df.dropna(subset=["Category"])
```

```python
pie = df["Category"].value_counts().rename_axis("Kategori").reset_index(name="Total")
```

```python
pie
```

```python
plt.figure(figsize=(10,10))
plt.title("Kategori dengan Persentase Jumlah Appnya ")
plt.pie(df["Category"].value_counts(), labels=pie["Kategori"], autopct='%.1f%%')
plt.show()
```

Hasil pada grafik pie di atas menunjukkan bahwa App dengan Kategori Family memiliki distribusi paling banyak sebesar 19% sedangkan Kategori Beauty memiliki distribusi paling sedikit sebesar 0,5%.

# 3. Grafik Scatter Plot

Scatter Plot atau Plot Sebar yaitu untuk menunjukkan hubungan antara dua variabel yang dibangun dari sumbu x dan y. Grafik di bawah menunjukkan persebaran data Type App Berdasarkan Rating App

```python
df["Type"].value_counts()
```

```python
sns.scatterplot(data = df, x="Type", y = "Rating", color="red")
plt.title("Persebaran Data Type App Berdasarkan Rating App")
plt.show()
```

Type App hanya ada dua yaitu Free(gratis) dan Paid(berbayar) dan grafik di atas adalah hasil persebarannya berdasarkan Rating App

# 4. Grafik Box Plot

Box Plot yaitu digunakan untuk melihat persebaran data dan mencari outliers. Di bawah ini dapat dilihat untuk persebaran data dan outliers pada data Rating App berdasarkan Type App 

```python
sns.boxplot(x= df["Rating"], y=df["Type"])
plt.title("Persebaran Data Rating App Berdasarkan Type App")
plt.show()
```

Grafik di atas menunjukkan Persebaran data pada Rating App berdasarkan Type App(Free dan Paid)
